```python
## dat is the read in data
afgh_c = natl_cases[natl_cases['Country/Region'] == 'Afghanistan'].drop(["Country/Region"], axis=1).unstack()
## convert from tuple to array
afgh_c = np.asarray(afgh_c)
## get case counts instead of cumulative counts
afgh_c = afgh_c[1 : afgh_c.size] - afgh_c[0 : (afgh_c.size - 1)]
## get the first non zero entry
afgh_c =  afgh_c[np.min(np.where(afgh_c !=  0)) : afgh_c.size]
plt.plot(afgh_c)
n = afgh_c.size
t = np.arange(0, n, 1)
lowess = sm.nonparametric.lowess
afgh_c_hat = lowess(afgh_c, t, frac=.05,return_sorted=False)
plt.plot(afgh_c)
plt.plot(afgh_c_hat)
```
